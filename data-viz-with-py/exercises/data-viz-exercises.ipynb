### Setting Up Your Data

Before you dive into data visualizations, you need to set up your dataframe and make sure that it is cleaned and ready to go.

```python
# import all librarys and modules needed
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("./Popular_Spotify_Songs.csv", encoding='latin-1')

df.head()

```

```python
# import and clean data
# display(df.isnull().count())
display(df.duplicated().count)
```

### Creating Visualizations

Now that the data is ready, you can use the dataset to create a number of different visualizations. You do not need to make the visualizations perfect; this is just for practice.

#### Line Chart

```python
# Single line chart
# plt.plot(df['released_year'],df['released_month'])

plt.plot(df['released_year'])
plt.show()
```

```python
# Multiple Line chart 
plt.plot(df['released_year'],df['released_month'])

plt.show()
```

#### Bar Chart

```python
# Vertical bar chart
plt.bar(df['released_year'],df['released_month'])
```

```python
# Horizontal bar chart
sb.barplot( y=df['released_month'],x=df['released_year'],orient="y")
```

```python
# Grouped bar chart
# sb.barplot( y=df['released_month'],x=df['released_year'],orient="y")

sb.barplot(y="released_day",x="released_month",hue="released_year",data=df)
# sns.barplot(x='Day', y='Customers', hue='Time', data=df) 
```

#### Scatter Plot

```python
# scatter plot
plt.scatter(data=df,y="released_day",x="released_month")
```

#### Histogram

```python
# histogram
plt.hist(data=df,x="released_month")
```

#### Pie Chart

```python
# pie chart
# plt.pie(data=df,x="released_month",labels="released_day")
plt.pie(df['artist_count'],labels=df["released_year"])
```
